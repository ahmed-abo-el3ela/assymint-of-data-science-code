#### 2. Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
x = [1, 2, 3, 4, 5]
y1 = [3,4,5,3,1]
y2 = [5,3,1,3,5]

plt.plot(x, y1)
plt.plot(x, y2)

plt.title('Plot of Two Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### 4. Write a Python program to plot two or more lines with different styles.


```python
x = [1, 2, 3, 4, 5]
y1 = [3,4,5,3,1]
y2 = [5,3,1,3,5]

plt.plot(x, y1 ,color='b', linestyle='--')
plt.plot(x, y2,color='r', linestyle=':')

plt.title('Plot of Two Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### 5. Write a Python program to plot two or more lines and set the line markers.


```python
x = [1, 2, 3, 4, 5]
y1 = [3,4,5,3,1]
y2 = [5,3,1,3,5]

plt.plot(x, y1 ,color='b', linestyle='--',marker='o')
plt.plot(x, y2,color='r', linestyle=':', marker='o')

plt.title('Plot of Two Functions')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### 6. Write a Python programming to display a bar chart of the popularity of programming Languages.
    Sample data:
    Programming languages: Java, Python, PHP, JavaScript, C#, C++
    Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7



```python
Programming_languages=[ 'Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
Popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(Programming_languages,Popularity)
plt.show()
```
